```python
from langchain.llms import OpenAI
from langchain.chat_models import ChatOpenAI
from langchain.prompts import ChatPromptTemplate, PromptTemplate, SystemMessagePromptTemplate, HumanMessagePromptTemplate,AIMessagePromptTemplate
from langchain.schema import AIMessage, HumanMessage, SystemMessage
import os
```

```python
api_key = os.environ["OPENAI_API_KEY"]
llm = OpenAI(openai_api_key=api_key, temperature=0.7, model = 'gpt-4o_mini')
chat = ChatOpenAI(openai_api_key=api_key, temperature=0.7, model = 'gpt-4o-mini')
```

```python
SystemMessage = 'You are a AI assistant that helps employee with their benefits questions. You are helpful, creative, clever, and very friendly.'
```

```python
system_message_prompt = SystemMessagePromptTemplate.from_template(SystemMessage)
```

```python
HumanMessage = 'What are the benefits of working at Acme Corporation?'
```

```python
human_message_prompt = HumanMessagePromptTemplate.from_template(HumanMessage)
```

```python
chat_prompt = ChatPromptTemplate.from_messages([system_message_prompt, human_message_prompt])
```

```python
## mini project
def travel_idea(interest, budget):
    # DEFINE SYSTEM TEMPLATE AND HUMAN MESSAGE WITH VARIABLES FOR INTEREST AND BUDGET
    system_message = 'You are a travel assistant that helps users find travel ideas based on their interests and budget. You are helpful, creative, clever, and very friendly.'
    human_message = 'I am interested in {interest} and my budget is {budget}. Can you suggest some travel ideas for me?'
    
    #DEFINE SYSTEM ANDHUMAN PROMPT TEMPLATES AND COMBINE THEM INTO A CHAT PROMPT
    system_message_prompt = SystemMessagePromptTemplate.from_template(system_message)
    human_message_prompt = HumanMessagePromptTemplate.from_template(human_message)
    
    chat_prompt = ChatPromptTemplate.from_messages([system_message_prompt, human_message_prompt])
    
    #INSERT THE VARIABLES INTO THE CHAT PROMPT
    response = chat(chat_prompt.format_messages(interest=interest, budget=budget))
    
    #RETURN THE RESPONSE CONTENT
    return response.content
```

## Take away
we can set up variables in system_message and human_message. And then set system message prompt and human_message_promt to set up templates and path the from_message function to path system message prompt and human message prompt. Lastly, return response as content. 

```python
print(travel_idea('hiking', '$1000'))
```

```python
# 2nd Prompt Test
# AI BOT LEGAL FOR SIMPLIFY EXPLAN LEGAL TERMS TO USER
def legal_explanation(legal_term):
    #DEFINE SYSTEM TEMPLATE , HUMAN MESSAGE AND AI MESSAGE WITH VARIABLE FOR LEGAL TERM
    system_message = 'You are a legal assistant that helps users understand legal terms. You are helpful, creative, clever, and very friendly.'
    human_message = 'Can you explain the legal term {legal_term} in simple language?'
    ai_message = 'Sure! I will explain it in simple, easy-to-understand language without complicated jargon.'


    #DEFINE SYSTEM, HUMAN AND AI PROMPT TEMPLATES AND COMBINE THEM INTO A CHAT PROMPT
    system_message_prompt = SystemMessagePromptTemplate.from_template(system_message)
    human_message_prompt = HumanMessagePromptTemplate.from_template(human_message)
    ai_message_prompt = AIMessagePromptTemplate.from_template(ai_message)


    chat_prompt = ChatPromptTemplate.from_messages([system_message_prompt, human_message_prompt, ai_message_prompt])
    
    response = chat(chat_prompt.format_messages(legal_term=legal_term))
    
    return response.content
```

```python
print(legal_explanation("habeas corpus"))
```
